```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.svm import SVC
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
le = LabelEncoder()
for i in range(3):
    df.iloc[:,i] = le.fit_transform(df.iloc[:,i])

i = 4
while i <= 34:
    df.iloc[:,i] = le.fit_transform(df.iloc[:,i])
    i = i+1
    
df.head()
```

```python
x = df.iloc[:,0:35]
y = df["Attrition"]
x = x.drop("Attrition", axis = 1)
x.head(),y.head()
```

```python
# Create decision tree classifer object
clf = RandomForestClassifier(random_state=0, n_jobs=-1)

# Train model
rtcmodel = clf.fit(x, y)
```

```python
cross_val_score(rtcmodel, x, y, scoring="accuracy"), cross_val_score(rtcmodel, x, y, scoring="f1"), cross_val_score(rtcmodel, x, y, scoring="precision"), cross_val_score(rtcmodel, x, y, scoring="recall")
```

```python
svcmodel = SVC(probability = True)
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,test_size=0.3,random_state=9)
svcmodel.fit(Xtrain,Ytrain)
cross_val_score(svcmodel, x, y, scoring="accuracy"), cross_val_score(svcmodel, x, y, scoring="f1"), cross_val_score(svcmodel, x, y, scoring="precision"), cross_val_score(svcmodel, x, y, scoring="recall")
```
